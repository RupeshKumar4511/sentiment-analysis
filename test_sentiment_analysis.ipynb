```python
import numpy as np
import pandas as pd
import re # regular expression : used for pattern matching
from nltk.corpus import stopwords # nltk : natural language toolkit -> used with text feature
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
```

```python
import nltk
nltk.download('stopwords')
```

```python
data  = pd.read_csv('/content/sma.posts.csv')
data.head()
index = data['_id']
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content) # it will remove letters which are not alphabet.
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split() # contents into list of words
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content) # list of words into content
  return stemmed_content
```

```python
data['stemmed_content'] = data['body'].apply(stemming)
```

```python
data['stemmed_content']
```

```python
vectorizer = pickle.load(open("/content/tfidf_vectorizer.pkl", "rb"))
```

```python
x_new = vectorizer.transform(data['stemmed_content'])
x_new
```

```python
# loading the saved model
filename = '/content/trained_model.sav'
loaded_model = pickle.load(open(filename,'rb'))
```

```python
index = pd.DataFrame(index)
```

```python
prediction = loaded_model.predict(x_new)
index['prediction']= pd.Series(prediction,name='prediction')
print(index[index['prediction']==1]['_id'].to_list())
```
